```python
############----------------river_rubbish-------------------

import os
from shutil import copyfile
import json

data_root = '/root/jinyfeng/datas/xining-proj-datas/floatingobjectsinriver'

label_savefolder = os.path.join(data_root, 'annotations')
if not os.path.exists(label_savefolder):
    os.makedirs(label_savefolder)
train_imgfolder = os.path.join(data_root, 'train')
val_imgfolder = os.path.join(data_root, 'val')
test_imgfolder = os.path.join(data_root, 'test')
if not os.path.exists(train_imgfolder):
    os.makedirs(train_imgfolder)
if not os.path.exists(val_imgfolder):
    os.makedirs(val_imgfolder)
if not os.path.exists(test_imgfolder):
    os.makedirs(test_imgfolder)
    
imagepath = os.path.join(data_root, 'JPEGImages')
labelpath = os.path.join(data_root, 'Annotations')

train_txtfile = os.path.join(data_root, 'ImageSets', 'Main', 'train.txt')
val_txtfile = os.path.join(data_root, 'ImageSets', 'Main', 'val.txt')
test_txtfile = os.path.join(data_root, 'ImageSets', 'Main', 'test.txt')

train_file = open(train_txtfile)
train_file_lines = train_file.readlines()
val_file = open(val_txtfile)
val_file_lines = val_file.readlines()
test_file = open(test_txtfile)
test_file_lines = test_file.readlines()
print(len(train_file_lines),len(val_file_lines),len(test_file_lines))

trainfile_json= os.path.join(label_savefolder, 'train.json')
valfile_json= os.path.join(label_savefolder, 'val.json')
testfile_json= os.path.join(label_savefolder, 'test.json')
    
train_json = open(trainfile_json, 'w')
val_json = open(valfile_json, 'w')
test_json = open(testfile_json, 'w')
from xml.dom.minidom import parse
import xml.dom.minidom
image_info_list_train=[]
image_annos_list_train=[]
image_id_list_train=[]
cate_list=[]
count_id = 10000
image_id = 30000
for imgfile in train_file_lines:
#     print(imgfile)
    imgfile = imgfile.split('\n')[0]
    source = os.path.join(imagepath, imgfile+'.jpg')
    target = os.path.join(train_imgfolder, imgfile+'.jpg')
    copyfile(source, target)
    label_file = os.path.join(labelpath, imgfile+'.xml')
#     print(label_file)
    DOMTree = xml.dom.minidom.parse(label_file)
    collection = DOMTree.documentElement
    
#     anno_info = collection.getAttribute("annotation")
    filename = collection.getElementsByTagName("filename")
    filename_val = filename[0].childNodes[0].data
#     print(filename_val)
    image_id += 1
    image_id_list_train.append(image_id)
    
    width = collection.getElementsByTagName("width")
    width_val = width[0].childNodes[0].data
    height = collection.getElementsByTagName("height")
    height_val = height[0].childNodes[0].data
#     print(image_id, width_val, height_val)
    image_dict={
        'file_name': imgfile+'.jpg',
        'height': int(height_val),
        'width': int(width_val),
        "id": int(image_id)
    }
    image_info_list_train.append(image_dict)
    
    object_info = collection.getElementsByTagName("object")
#     print(len(object_info))
    for obj in object_info:
        name = obj.getElementsByTagName("name")
        name_str = name[0].childNodes[0].data
#         print(name_str)
        if 'leaf' not in name_str and \
            'milk-box' not in name_str and \
            'plastic-garbage' not in name_str and \
            'bottle' not in name_str and \
            'branch' not in name_str and \
            'ball' not in name_str and \
            'grass' not in name_str and \
            'plastic-bag' not in name_str:
            print(name_str, imgfile)
        if name_str=='leaf' or name_str=='milk-box' or \
            name_str=='plastic-garbage' or name_str=='bottle' or \
            name_str=='branch' or name_str=='ball' or \
            name_str=='grass' or name_str=='plastic-bag' :
            xmin=obj.getElementsByTagName("xmin")
            xmin_val = float(xmin[0].childNodes[0].data)
            ymin=obj.getElementsByTagName("ymin")
            ymin_val = float(ymin[0].childNodes[0].data)
            xmax=obj.getElementsByTagName("xmax")
            xmax_val = float(xmax[0].childNodes[0].data)
            ymax=obj.getElementsByTagName("ymax")
            ymax_val = float(ymax[0].childNodes[0].data)
#             print(xmin_val, ymin_val, xmax_val, ymax_val)
            if name_str=='leaf':
                category_id = 1
            elif name_str=='milk-box':
                category_id = 2
            elif name_str=='plastic-garbage':
                category_id = 3
            elif name_str=='bottle':
                category_id = 4
            elif name_str=='branch':
                category_id = 5
            elif name_str=='ball':
                category_id = 6
            elif name_str=='grass':
                category_id = 7
            elif name_str=='plastic-bag':
                category_id = 8
            else:
                print(name_str)
                category_id = 8
            count_id += 1
            width = xmax_val-xmin_val
            height = ymax_val-ymin_val
            anno_dict = {
                "iscrowd": 0,
                "area": height*width,
                "image_id": int(image_id),
                "bbox":[xmin_val, ymin_val, width, height],
                "category_id": category_id,
                "id": count_id
            }
            image_annos_list_train.append(anno_dict)

category_dict1={
    "supercategory": "rubbish",
    "id": 1,
    "name": "leaf"
}
category_dict2={
    "supercategory": "rubbish",
    "id": 2,
    "name": "milk-box"
}
category_dict3={
    "supercategory": "rubbish",
    "id": 3,
    "name": "plastic-garbage"
}
category_dict4={
    "supercategory": "rubbish",
    "id": 4,
    "name": "bottle"
}
category_dict5={
    "supercategory": "rubbish",
    "id": 5,
    "name": "branch"
}
category_dict6={
    "supercategory": "rubbish",
    "id": 6,
    "name": "ball"
}
category_dict7={
    "supercategory": "rubbish",
    "id": 7,
    "name": "grass"
}
category_dict8={
    "supercategory": "rubbish",
    "id": 8,
    "name": "plastic-bag"
}
cate_list.append(category_dict1)
cate_list.append(category_dict2)
cate_list.append(category_dict3)
cate_list.append(category_dict4)
cate_list.append(category_dict5)
cate_list.append(category_dict6)
cate_list.append(category_dict7)
cate_list.append(category_dict8)

train_label_json = json.dumps(dict(images=image_info_list_train,
                               annotations=image_annos_list_train,
                                  categories=cate_list))
train_json.writelines(train_label_json)
train_json.close()


image_info_list_val=[]
image_annos_list_val=[]
image_id_list_val=[]
for imgfile in val_file_lines:
#     print(imgfile)
    imgfile = imgfile.split('\n')[0]
    source = os.path.join(imagepath, imgfile+'.jpg')
    target = os.path.join(val_imgfolder, imgfile+'.jpg')
    copyfile(source, target)
    label_file = os.path.join(labelpath, imgfile+'.xml')
#     print(label_file)
    DOMTree = xml.dom.minidom.parse(label_file)
    collection = DOMTree.documentElement
    
#     anno_info = collection.getAttribute("annotation")
    filename = collection.getElementsByTagName("filename")
    filename_val = filename[0].childNodes[0].data
    #     print(filename_val)
    image_id += 1
    image_id_list_val.append(image_id)
    
    width = collection.getElementsByTagName("width")
    width_val = width[0].childNodes[0].data
    height = collection.getElementsByTagName("height")
    height_val = height[0].childNodes[0].data
#     print(image_id, width_val, height_val)
    image_dict={
        'file_name': imgfile+'.jpg',
        'height': int(height_val),
        'width': int(width_val),
        "id": int(image_id)
    }
    image_info_list_val.append(image_dict)
    
    object_info = collection.getElementsByTagName("object")
#     print(len(object_info))
    for obj in object_info:
        name = obj.getElementsByTagName("name")
        name_str = name[0].childNodes[0].data
#         print(name_str)
        if 'leaf' not in name_str and \
            'milk-box' not in name_str and \
            'plastic-garbage' not in name_str and \
            'bottle' not in name_str and \
            'branch' not in name_str and \
            'ball' not in name_str and \
            'grass' not in name_str and \
            'plastic-bag' not in name_str:
            print(name_str, imgfile)
        if name_str=='leaf' or name_str=='milk-box' or \
            name_str=='plastic-garbage' or name_str=='bottle' or \
            name_str=='branch' or name_str=='ball' or \
            name_str=='grass' or name_str=='plastic-bag' :
            xmin=obj.getElementsByTagName("xmin")
            xmin_val = float(xmin[0].childNodes[0].data)
            ymin=obj.getElementsByTagName("ymin")
            ymin_val = float(ymin[0].childNodes[0].data)
            xmax=obj.getElementsByTagName("xmax")
            xmax_val = float(xmax[0].childNodes[0].data)
            ymax=obj.getElementsByTagName("ymax")
            ymax_val = float(ymax[0].childNodes[0].data)
#             print(xmin_val, ymin_val, xmax_val, ymax_val)
            if name_str == 'leaf':
                category_id = 1
            elif name_str=='milk-box':
                category_id = 2
            elif name_str=='plastic-garbage':
                category_id = 3
            elif name_str=='bottle':
                category_id = 4
            elif name_str=='branch':
                category_id = 5
            elif name_str=='ball':
                category_id = 6
            elif name_str=='grass':
                category_id = 7
            elif name_str=='plastic-bag':
                category_id = 8
            else:
                print(name_str)
                category_id = 8
            count_id += 1
            width = xmax_val-xmin_val
            height = ymax_val-ymin_val
            anno_dict = {
                "iscrowd": 0,
                "area": height*width,
                "image_id": int(image_id),
                "bbox":[xmin_val, ymin_val, width, height],
                "category_id": category_id,
                "id": count_id
            }
            image_annos_list_val.append(anno_dict)
            
val_label_json = json.dumps(dict(images=image_info_list_val,
                               annotations=image_annos_list_val,
                                categories=cate_list))
val_json.writelines(val_label_json)
val_json.close()


image_info_list_test=[]
image_annos_list_test=[]
image_id_list_test=[]
for imgfile in test_file_lines:
#     print(imgfile)
    imgfile = imgfile.split('\n')[0]
    source = os.path.join(imagepath, imgfile+'.jpg')
    target = os.path.join(test_imgfolder, imgfile+'.jpg')
    copyfile(source, target)
    label_file = os.path.join(labelpath, imgfile+'.xml')
#     print(label_file)
    DOMTree = xml.dom.minidom.parse(label_file)
    collection = DOMTree.documentElement
    
#     anno_info = collection.getAttribute("annotation")
    filename = collection.getElementsByTagName("filename")
    filename_val = filename[0].childNodes[0].data
    #     print(filename_val)
    image_id += 1
    image_id_list_test.append(image_id)
    
    width = collection.getElementsByTagName("width")
    width_val = width[0].childNodes[0].data
    height = collection.getElementsByTagName("height")
    height_val = height[0].childNodes[0].data
#     print(image_id, width_val, height_val)
    image_dict={
        'file_name': imgfile+'.jpg',
        'height': int(height_val),
        'width': int(width_val),
        "id": int(image_id)
    }
    image_info_list_test.append(image_dict)
    
    object_info = collection.getElementsByTagName("object")
#     print(len(object_info))
    for obj in object_info:
        name = obj.getElementsByTagName("name")
        name_str = name[0].childNodes[0].data
#         print(name_str)
        if 'leaf' not in name_str and \
            'milk-box' not in name_str and \
            'plastic-garbage' not in name_str and \
            'bottle' not in name_str and \
            'branch' not in name_str and \
            'ball' not in name_str and \
            'grass' not in name_str and \
            'plastic-bag' not in name_str:
            print(name_str, imgfile)
        if name_str=='leaf' or name_str=='milk-box' or \
            name_str=='plastic-garbage' or name_str=='bottle' or \
            name_str=='branch' or name_str=='ball' or \
            name_str=='grass' or name_str=='plastic-bag' :
            xmin=obj.getElementsByTagName("xmin")
            xmin_val = float(xmin[0].childNodes[0].data)
            ymin=obj.getElementsByTagName("ymin")
            ymin_val = float(ymin[0].childNodes[0].data)
            xmax=obj.getElementsByTagName("xmax")
            xmax_val = float(xmax[0].childNodes[0].data)
            ymax=obj.getElementsByTagName("ymax")
            ymax_val = float(ymax[0].childNodes[0].data)
#             print(xmin_val, ymin_val, xmax_val, ymax_val)
            if name_str == 'leaf':
                category_id = 1
            elif name_str=='milk-box':
                category_id = 2
            elif name_str=='plastic-garbage':
                category_id = 3
            elif name_str=='bottle':
                category_id = 4
            elif name_str=='branch':
                category_id = 5
            elif name_str=='ball':
                category_id = 6
            elif name_str=='grass':
                category_id = 7
            elif name_str=='plastic-bag':
                category_id = 8
            else:
                print(name_str)
                category_id = 8
            count_id += 1
            width = xmax_val-xmin_val
            height = ymax_val-ymin_val
            anno_dict = {
                "iscrowd": 0,
                "area": height*width,
                "image_id": int(image_id),
                "bbox":[xmin_val, ymin_val, width, height],
                "category_id": category_id,
                "id": count_id
            }
            image_annos_list_test.append(anno_dict)
            
test_label_json = json.dumps(dict(images=image_info_list_test,
                               annotations=image_annos_list_test,
                                categories=cate_list))
test_json.writelines(test_label_json)
test_json.close()


print(len(image_id_list_train))
print(len(set(image_id_list_train)))
print(len(image_id_list_val))
print(len(set(image_id_list_val)))
print(len(image_id_list_test))
print(len(set(image_id_list_test)))
```

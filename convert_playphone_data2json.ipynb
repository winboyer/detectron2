```python
############----------------play_phone-------------------
import os
from shutil import copyfile
import json
import cv2

data_root = '/root/jinyfeng/datas/xining-proj-datas/play_phone-data'

label_savefolder = os.path.join(data_root, 'annotations')
if not os.path.exists(label_savefolder):
    os.makedirs(label_savefolder)
train_imgfolder = os.path.join(data_root, 'train')
val_imgfolder = os.path.join(data_root, 'val')
    
labelpath = os.path.join(data_root, 'labels')

train_filenames = os.listdir(train_imgfolder)
val_filenames = os.listdir(val_imgfolder)
print(len(train_filenames), len(val_filenames))
imglist_train = []
imglist_val = []
for filename in train_filenames:
    if os.path.splitext(filename)[1] == '.jpg' or os.path.splitext(filename)[1] == '.JPG':
        imglist_train.append(filename)
for filename in val_filenames:
    if os.path.splitext(filename)[1] == '.jpg' or os.path.splitext(filename)[1] == '.JPG':
        imglist_val.append(filename)
print(len(imglist_train), len(imglist_val))

image_info_list_train=[]
image_annos_list_train=[]
image_id_list_train=[]
cate_list=[]
image_info_list_val=[]
image_annos_list_val=[]
image_id_list_val=[]

count_id = 10000
image_id = 20000

category_id=1

trainfile_json= os.path.join(label_savefolder, 'train.json')
train_json = open(trainfile_json, 'w')
for imgfile in imglist_train:
#     print(imgfile)
    image = cv2.imread(os.path.join(train_imgfolder, imgfile))
#     print(image.shape)
    height, width, _ = image.shape
#     print(height, width)
    image_id += 1
    image_dict={
        'file_name': imgfile,
        'height': height,
        'width': width,
        "id": image_id
    }
    image_id_list_train.append(image_id)
    image_info_list_train.append(image_dict)
    
    name = imgfile.split(".")[0]
    label_file = os.path.join(labelpath, 'train', name+'.txt')
#     print(label_file)
    f = open(label_file)
    lines = f.readlines()
#     if len(lines)>1:
#         print(label_file, lines, len(lines))
    for line in lines:
#         print(label_file, line)
        label = int(line.split(' ')[0])
        center_x = width * float(line.split(' ')[1])
        center_y = height * float(line.split(' ')[2])
        rect_w = width * float(line.split(' ')[3])
        rect_h = height * float(line.split(' ')[4].split('\n')[0])
        x1 = int(center_x-rect_w/2+0.5)
        y1 = int(center_y-rect_h/2+0.5)
        x2 = int(center_x+rect_w/2+0.5)
        y2 = int(center_y+rect_h/2+0.5)
        if label != 0:
            print(label_file, line)
#         print(width, height, x1, y1, x2, y2)
        count_id += 1
        anno_dict = {
            "iscrowd": 0,
            "area": rect_w * rect_h,
            "image_id": int(image_id),
            "bbox":[x1, y1, rect_w, rect_h],
            "category_id": category_id,
            "id": count_id
        }
        image_annos_list_train.append(anno_dict)
        
category_dict1={
    "supercategory": "phone",
    "id": 1,
    "name": "play_phone"
}
cate_list.append(category_dict1)
train_label_json = json.dumps(dict(images=image_info_list_train,
                               annotations=image_annos_list_train,
                                  categories=cate_list))
# print(train_label_json)
train_json.writelines(train_label_json)
train_json.close()


valfile_json= os.path.join(label_savefolder, 'val.json')
val_json = open(valfile_json, 'w')
for imgfile in imglist_val:
#     print(imgfile)
    image = cv2.imread(os.path.join(val_imgfolder, imgfile))
#     print(image.shape)
    height, width, _ = image.shape
#     print(height, width)
    image_id += 1
    image_dict={
        'file_name': imgfile,
        'height': height,
        'width': width,
        "id": image_id
    }
    image_id_list_val.append(image_id)
    image_info_list_val.append(image_dict)
    
    name = imgfile.split(".")[0]
    label_file = os.path.join(labelpath, 'val', name+'.txt')
#     print(label_file)
    f = open(label_file)
    lines = f.readlines()
#     if len(lines)>1:
#         print(label_file, lines, len(lines))
    for line in lines:
#         print(label_file, line)
        label = int(line.split(' ')[0])
        center_x = width * float(line.split(' ')[1])
        center_y = height * float(line.split(' ')[2])
        rect_w = width * float(line.split(' ')[3])
        rect_h = height * float(line.split(' ')[4].split('\n')[0])
        x1 = int(center_x-rect_w/2+0.5)
        y1 = int(center_y-rect_h/2+0.5)
        x2 = int(center_x+rect_w/2+0.5)
        y2 = int(center_y+rect_h/2+0.5)
        if label != 0:
            print(label_file, line)
#         print(width, height, x1, y1, x2, y2)
        count_id += 1
        anno_dict = {
            "iscrowd": 0,
            "area": rect_w * rect_h,
            "image_id": int(image_id),
            "bbox":[x1, y1, rect_w, rect_h],
            "category_id": category_id,
            "id": count_id
        }
        image_annos_list_val.append(anno_dict)

val_label_json = json.dumps(dict(images=image_info_list_val,
                               annotations=image_annos_list_val,
                                  categories=cate_list))
# print(val_label_json)
val_json.writelines(val_label_json)
val_json.close()


print(len(image_id_list_train))
print(len(set(image_id_list_train)))
print(len(image_id_list_val))
print(len(set(image_id_list_val)))
```
